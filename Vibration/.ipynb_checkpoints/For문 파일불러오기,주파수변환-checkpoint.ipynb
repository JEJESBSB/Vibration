- dir_path = 특정 용량 폴더로 이동
- files = 특정 불량 폴더로 이동 후 불러올 데이터 갯수 설정함 
- for문 => 불러온 파일 하나의 데이터프레임으로 합치기 

```python
def read_files(kw,defect):
    dir_path = "/Volumes/ExternalSSD/Training/vibration/"+kw+'kW/' 
    files = glob.glob(os.path.join(dir_path+'/*/'+defect+'/','*.*'))[0:10]
    pieces = []
    for i in range(len(files)):
        file = files[i]
#       print(i,file)
        frame = pd.read_csv(file,skiprows=8)
#       print(frame)
        pieces.append(frame)
    df = pd.concat(pieces,ignore_index=True)
    df.drop(columns=['Unnamed: 2'],axis=1,inplace=True)
    df.columns=['time','data']
    return df
```

```python
df1 = read_files('3.7','회전체불평형') 
```

```python
# 처음 3초만 확인
df1 = df1[:12000]
```

```python
df1['data'].plot(figsize=(20,3))
```

- n = len(df1['data']) 
- k = np.arange(n)
- Fs : 1/0.00025 # 1초에 반복되는 횟수, sample interval time(0.00025초) 의 역수를 통해 sample frequency 계산 => 4000
- T : n/Fs # end of time => 3
- freq = k/T # 주파수 영역의 간격 {freq = freq[range(int(n/2))]}

```python
def fft(df):
    n = len(df['data']) 
    k = np.arange(n)
    Fs = 1/0.00025 # 1초에 반복되는 횟수 
    T = n/Fs # end of time 
    freq = k/T
    freq = freq[range(int(n/2))]
    Y = np.fft.fft(df['data']) # df1의 시간영역의 데이터에서 주파수 성분 찾게됨 
    Y = Y[range(int(n/2))]/n
    
    fig, ax = plt.subplots(2, 1, figsize=(12,8))
    ax[0].plot(df['time'].iloc[:4000],df['data'].iloc[:4000])
    ax[0].set_xlabel('Time')
    ax[0].set_ylabel('Amplitude'); ax[0].grid(True)
    ax[1].plot(freq, abs(Y), 'r', linestyle=' ', marker='^') 
    ax[1].set_xlabel('Freq (Hz)')
    ax[1].set_ylabel('|Y(freq)|')
    ax[1].vlines(freq, [0], abs(Y))
    ax[1].set_xlim([0, 1000]); ax[1].grid(True)
    plt.show()
```

```python
fft(df1)
```
